### <center>Q no 1(b)</center>

```python
Array = np.random.randint(99,size=(9))
print(Array)
def bubbleSort(Array):
    i = 0
    while i<len(Array):
        j = 0
        while j<len(Array)-1:
            if Array[j+1] < Array[j]:
                Array[j], Array[j+1] = Array[j+1], Array[j]
            j += 1
        i += 1
    return Array
print(bubbleSort(Array))
```

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,size=10000)
ax = sns.distplot(data_binom,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')

print("Now this is the actual binomial distribution of data...\nand we can see that if we increase right tale or upper boundary to the length of lower boundary\nor left tale and close the pillars of the data or compress our data we will have our data in\nnormal distribution form which is like if we decrease the contant value p from 0.8 to 0.5 and\nincrease the value of n then we will nearly have our data shown in Normal Distributed Form\nwhereas we are using binomial built-in method...\n\nLet's try that")
```

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10000,p=0.5,size=10000)
ax = sns.distplot(data_binom,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution', ylabel='Frequency')

print("That's all...!!!")
```
